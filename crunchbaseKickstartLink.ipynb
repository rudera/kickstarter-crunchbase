```python
for index, row in ks_df.iterrows():
    if ks_df.at[index, 'twitter'] != '[]':
        ks_df.at[index, 'twitter'] = eval(ks_df.at[index,'twitter'])[0].split('//')[1]
    if ks_df.at[index, 'facebook'] != '[]':
        ks_df.at[index,'facebook'] = eval(ks_df.at[index,'facebook'])[0].split('//')[1]
ks_df
```

```python
for index, row in cb_df.iterrows():
    if pd.isna(cb_df.at[index, 'twitter']):
        continue
    elif len(cb_df.at[index, 'twitter'].split('//')) < 2:
        continue
    else:
        cb_df.at[index, 'twitter'] = cb_df.at[index, 'twitter'].split('//')[1]
for index, row in cb_df.iterrows():
    if pd.isna(cb_df.at[index, 'facebook']):
        continue
    elif len(cb_df.at[index, 'facebook'].split('//')) < 2:
        continue
    else:
        cb_df.at[index, 'facebook'] = cb_df.at[index, 'facebook'].split('//')[1]
cb_df
```

```python
tw_merge= pd.merge(left=cb_df, right=ks_df, left_on='twitter', right_on='twitter')
tw_merge
```

```python
tw_merge = tw_merge.drop_duplicates(subset=['_id'],ignore_index=True)
tw_merge
```
